```python
import pandas as pd
from google.colab import files

# Subimos el archivo
uploaded = files.upload()

# Obtenemos el nombre del archivo
file_name = list(uploaded.keys())[0]

# Cargamos solo la hoja "VENCIMIENTOS GENERALES"
vencimientos_generales = pd.read_excel(file_name, sheet_name="VENCIMIENTOS GENERALES", engine='openpyxl')

# Exploramos las primeras filas para verificar el formato de la fecha
print(vencimientos_generales.head())


vencimientos_generales['FECHA '] = pd.to_datetime(vencimientos_generales['FECHA '], errors='coerce')

# Agrupamos los datos por fecha y contamos cuántos radicados hay por día
radicados_por_dia = vencimientos_generales.groupby(vencimientos_generales['FECHA '].dt.date).size()

# Mostramos el resultado
print(radicados_por_dia)
```
